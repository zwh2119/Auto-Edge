```python
import matplotlib.pyplot as plt
import os
import json
```

```python
work_path = 'resource_data/resource1.json'
```

```python
with open(work_path, 'r') as f:
    data = json.load(f)

use_bandwidth = []
total_bandwidth = []
for resource in data:
    use_bandwidth.append(resource['total_bandwidth'])
    total_bandwidth.append(resource['bandwidth'])
```

```python
import numpy as np

color = ['#2878b5',
'#c82423']

plt.figure(figsize=(10, 4))
plt.xlabel('time', fontsize=15)
plt.ylabel('bandwidth', fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.plot((total_bandwidth), color=color[0], linewidth=2, label='available bandwidth')
plt.plot((use_bandwidth), color=color[1], linewidth=2, label='occupies bandwidth')


# plt.ylim(0, 0.14)
plt.legend()
```
